```python
# Question 4: Find the maximum value in tuple (23, 45, 12, 67, 34, 89, 56)
print("Question 4: Find the maximum value in tuple (23, 45, 12, 67, 34, 89, 56)")

tuple1=(23, 45, 12, 67, 34, 89, 56)
max_value=0
for i in tuple1:
    if i>max_value:
        max_value=i
print(f"the maximum value in tuple '{max_value}'")
```

```python
# Question 10: Find the average of numbers in tuple (15, 23, 31, 42, 56, 78)
print("Question 10: Find the average of numbers in tuple (15, 23, 31, 42, 56, 78)")

tuple1=(15, 23, 31, 42, 56, 78)
length=len(tuple1)
add=0
for i in tuple1:
    add=add+i
   
print(f"{add/length:.2f}")

```
