```python
sorted_num = [21, 23, 34, 43, 45, 46, 66, 67, 87]
freq = {}

# Count frequencies
for i in sorted_num:
    freq[i] = freq.get(i, 0) + 1

print(freq)

# Find maximum frequency
max_values = max(freq.values())
if max_values > 1:   # Only check if there's a repetition
    for key in freq:
        if freq[key] == max_values:
            print(f"{key} occurs {max_values} times")
else:
    print("No frequence")
```
