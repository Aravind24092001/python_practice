```python
print("Question 23: Convert 'hello world' to 'hElLo WoRlD' (alternating case)")

s1 = 'hello world'
result = ""
letter_index = 0
for ch in s1:
    if ch.isalpha():
        if letter_index % 2 == 0:
            result += ch.lower()
        else:
            result += ch.upper()
        letter_index += 1
    else:
        result += ch

print(result)
```

```python
print("Question 31: Check if string is a valid email format: 'user@example.com'")

user_input = "user@example.com"

parts = user_input.split('@')

if len(parts) != 2:
    print("Invalid email")
else:
    username, domain = parts

    if username == "" or domain == "":
        print("Invalid email")
    elif '.' not in domain:
        print("Invalid email")
    else:
        domain_parts = domain.split('.')
        if "" in domain_parts:
            print("Invalid email")
        else:
            print("Valid email")
```

```python
print("Question 46: Convert 'hello world' to Pig Latin")
# Your code here
'''
For each word:
1️⃣ If a word starts with a vowel → add "way" at the end
2️⃣ If a word starts with a consonant →
    • move the first letter to the end
    • add "ay"
'''
sentence = 'hello world'
arr = sentence.split()
result = []

vowels = "aeiou"

for word in arr:
    if word[0].lower() in vowels:
        result.append(word + "way")
    else:
        result.append(word[1:] + word[0] + "ay")

print(" ".join(result))
```
