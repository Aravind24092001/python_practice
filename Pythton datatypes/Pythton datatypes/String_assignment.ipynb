```python
# Question: Count vowels in the string "Hello World"
print("SOLVED EXAMPLE:")
print("Count vowels in the string 'Hello World'")
text = "Hello World"
vowels = "aeiouAEIOU"
count = sum(1 for char in text if char in vowels)
print(f"String: {text}")
print(f"Number of vowels: {count}")
print("-" * 50)
```

```python
# Question 2: Check if "racecar" is a palindrome
print("\nQuestion 2: Check if 'racecar' is a palindrome")

palindrome="racecar"
reverse_car=palindrome[::-1]
if (palindrome==reverse_car):
    print(f"is a palindrome :'{palindrome}'")
else:
    print(f"is not a palindrome: '{palindrome}'")
```

```python
# Question 12: Check if two strings are anagrams: "listen" and "silent"
print("Question 12: Check if two strings are anagrams: 'listen' and 'silent'")

String1='listen'
String2='silent'
print(sorted(String1))
if sorted(String1) == sorted(String2):
  print(f'{String1} and {String2} are anagrams.')
else:
  print(f'{String1} and {String2} are not anagrams')
```

```python
# Question 14: Count consonants in "Hello World"
print("Question 14: Count consonants in 'Hello World'")

S1='Hello World'
S2="AEIOUaeiou"
count=0

for i in S1:
    if i.isalpha() and i not in S2:
        count+=1

print(count)
```
