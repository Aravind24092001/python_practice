```python
# Question 4: Sort the list [64, 34, 25, 12, 22, 11, 90] in descending order
print("Question 4: Sort the list [64, 34, 25, 12, 22, 11, 90] in descending order")

list1 = [64, 34, 25, 12, 22, 11, 90]
n = len(list1)

for i in range(n):
    for j in range(0, n - i - 1):
        if list1[j] < list1[j + 1]:
            
            list1[j], list1[j + 1] = list1[j + 1], list1[j]

print("Sorted list (Descending):", list1)
```

```python
# Question 34: Count the depth of nesting in [1, [2, 3], [4, [5, 6]], 7]
print("Question 34: Count the depth of nesting in [1, [2, 3], [4, [5, 6]], 7]")
nested_list = [1, [2, 3], [4, [5, 6]], 7]

level = [nested_list]
depth = 0

while len(level) != 0:
    next_level = []

    for item in level:
        if type(item) == list:
            depth = depth + 1
            for x in item:
                if type(x) == list:
                    next_level.append(x)

    level = next_level

print("Depth of nesting:", depth)
```
