```python
print("1. Count how many vowels are in a given string.")

String1="programming"
vowels="AEIOUaeiou"
count=0
for i in String1:
    if i in vowels:
        count+=1

print(count)
```

```python
#2. Reverse a number using only a for loop (no slicing).
#Example: Input → 12345, Output → 54321

string = input("enter string to reverse: ")
l = len(string)

for i in range(l,0,-1):
    print(string[i-1], end="")
```

```python
#5. Find the factorial of a number.

number=5
factorial=1

while number>0:

    factorial=factorial*number
    number=number-1
    
print(f" factorial of {number} is a {factorial}")




```
