```python
def calculate_average(scores):
    total=sum(scores)
    avg=total/len(scores)
    return(avg)
```

```python
calculate_average([24,16,28])
```

```python
def calculating_sum(*number):
    total=0
    for num in number:
        total+=num
    return(total)
```

```python
calculating_sum(2,3,4,5,6,7)
```

```python
def student_details(**details):
    print("student info")
    for key,value in details.items():
        print(f"{key} {value}")
```

```python
student_details(name="Aravind",age="24",city="Anantapur")
```

```python
def complex_function(a,b,*args,c=10,**kwargs):
    print(f"a (position) {a}")
    print(f"b (position) {b}")
    print(f"args (variable position): {args}")
    print(f" default: {c}")
    print(f"*kwargs (variable keyword): {kwargs}")
```

```python
complex_function(3,4,3,4,5, c=20, x=100, y = 200)
```

```python
def calculate_starts(numbers):
    minimum=min(numbers)
    maximum=max(numbers)
    average=sum(numbers)/len(numbers)

    return minimum, maximum, average
```

```python
min_val, max_val, avg_val = calculate_starts([10,20,30,40,50,60])
```
