```python
numbers=[2,3,4,5,6,7,8]
squard_list=[]

for num in numbers:
    squard_list.append(num**2)

print(f"number_lists: {numbers}")
print(f"squard_lists: {squard_list}")
```

```python
squard1_lists=[num**2  for num in numbers if num%2==0 ]
print(squard1_lists)
```
